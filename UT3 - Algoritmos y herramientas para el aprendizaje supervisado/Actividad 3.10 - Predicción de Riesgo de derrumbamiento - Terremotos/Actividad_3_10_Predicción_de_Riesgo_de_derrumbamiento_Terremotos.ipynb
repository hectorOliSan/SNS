## Definición detallada del problema o reto:

#### **Descripción del problema**:

**El objetivo del problema es predecir el grado de daño causado por un terremoto en un edificio determinado**. La variable objetivo es `damage_grade`, la cual es una variable ordinal que se compone de tres categorías diferentes que representan diferentes niveles de daño. Estos niveles son:

- `1` Daño bajo
- `2` Cantidad media de daño
- `3` Destrucción casi completa

#### **Características**:

El conjunto de datos consiste principalmente en información sobre la estructura de los edificios y su propiedad legal. Cada fila del conjunto de datos representa un edificio específico en la región afectada por el [terremoto de Gorkha](https://es.wikipedia.org/wiki/Terremoto_de_Nepal_de_abril_de_2015).

Hay 39 columnas en este conjunto de datos, donde la columna **building_id** es un identificador único y aleatorio. Las 38 funciones restantes se describen en la siguiente sección. Las variables categóricas se cambiaron por caracteres ascii aleatorios en minúsculas. La aparición del mismo carácter en columnas distintas no implica el mismo valor original.

### Transformación de Datos

#### Conversión, de Categóricos a Numéricos si procede

Es común que los datos se presenten en diferentes formatos, como valores categóricos o numéricos. En algunos casos, los algoritmos de aprendizaje automático solo pueden trabajar con datos numéricos, lo que significa que es necesario convertir los valores categóricos en números.

La conversión de datos categóricos a numéricos implica asignar un número a cada categoría única. Esto permite que los algoritmos de aprendizaje automático trabajen con los datos de manera más efectiva y precisa. Además, es común que los algoritmos de aprendizaje automático requieran que los datos se presenten en formato float, lo que significa que los valores deben tener un decimal.

```python
# Lista de valores categóricos distintos del Dataframe de train
categoricos = [
  'land_surface_condition', 'foundation_type', 'roof_type','ground_floor_type',
  'other_floor_type', 'position','plan_configuration', 'legal_ownership_status'
]
```

```python
# Ver la conversión de los valores categóricos
def ver_conversion(df, categoricos):
  for c in categoricos:
    print(c + ':')
    print(list(df[c].unique()))
    print(list(range(len(df[c].unique()))),)
    print()
```

```python
# Ver Conversiones del Dataframe de train
ver_conversion(df_train, categoricos)
```

```python
# Ver Conversiones del Dataframe de test
ver_conversion(df_test, categoricos)
```

```python
# Conversión, de Categóricos a Numéricos si procede
def conversion(df, categoricos):
  for c in categoricos:
    df[c] = df[c].replace(
      list(df[c].unique()), 
      list(range(len(df[c].unique())))
    )
    df[c] = df[c].astype('int64')
```

```python
# Conversiones del Dataframe de train
conversion(df_train, categoricos)
```

```python
# Conversiones del Dataframe de test
conversion(df_test, categoricos)
```

#### Escalado

```python
# Crea un objeto MaxAbsScaler

scaler = MaxAbsScaler()

# Ajuste y transformación de los datos de entrenamiento y prueba

df_train_e = pd.DataFrame(columns=df_train.columns)
df_test_e = pd.DataFrame(columns=df_test.columns)

df_train_e[df_train.columns] = scaler.fit_transform(df_train)
df_test_e[df_test.columns] = scaler.transform(df_test)
```

El escalado absoluto máximo, **MaxAbsScaler**: es un método de escalado utilizado en el preprocesamiento de datos en el aprendizaje automático. La idea detrás de MaxAbsScaler es escalar los datos de modo que el valor absoluto más grande de cada característica sea 1.

MaxAbsScaler es útil cuando se trabaja con datos que tienen valores extremadamente altos en algunas características y valores muy bajos en otras. Al escalar los datos, se pueden evitar problemas relacionados con la diferencia de magnitudes entre las características.

Los valores de correlación más altos se representan en un color más oscuro, mientras que los valores de correlación más bajos se representan en un color más claro. El mapa de calor es útil porque permite visualizar rápidamente las relaciones entre las variables y resaltar patrones que pueden ser difíciles de detectar en una tabla de números.

Además, el mapa de calor es una forma eficiente de identificar variables altamente correlacionadas, lo que puede ser útil en la selección de variables para el análisis o modelado. Si dos variables están altamente correlacionadas, es posible que solo se necesite una de ellas en el modelo para explicar la variabilidad de los datos. Esto puede ayudar a simplificar el modelo y reducir la complejidad.

```python
from scipy.cluster import hierarchy

# Matriz de distancia
d = hierarchy.distance.pdist(df_train_e.T)
avD = np.average(d)
print("Distancia Media:{:6.2f}".format(avD))

# Calculo del dendrograma
Z = hierarchy.linkage(d, 'ward')

# Dibujo del dendrograma
plt.figure(figsize=(10, 5))
labels = df_train_e.columns.to_list()
dn = hierarchy.dendrogram(Z, color_threshold=avD, labels=labels, leaf_rotation=90)
plt.show()
```

Un **dendrograma** es una representación gráfica de las similitudes y diferencias entre diferentes objetos/características o grupos de objetos/características. Los objetos/características o grupos se muestran como nodos, y las conexiones entre ellos representan la similitud o distancia entre ellos. Es un concepto totalmente diferente a la correlación de variables.

Es un tipo de diagrama en forma de árbol utilizado en análisis de clustering ( Crear grupos en aprendizaje no supervisado que veremos más adelante), pero que lo podemos utilizar para analizar las similitudes/cercanía entre características.

## Entrenamiento

### Modelo 1: Regresión

**Logistic Regression (Regresión Logística)** es un algoritmo de aprendizaje supervisado utilizado en problemas de clasificación, en el cual el objetivo es predecir la probabilidad de que una observación pertenezca a una determinada clase. A diferencia de la regresión lineal, la regresión logística tiene como objetivo predecir una variable categórica en lugar de una variable numérica continua.

La regresión logística utiliza una función logística para predecir la probabilidad de que una observación pertenezca a una clase determinada. Esta función toma como entrada una combinación lineal de los atributos de entrada (variables independientes) y los coeficientes del modelo, y devuelve una probabilidad entre 0 y 1. Si la probabilidad es mayor que un umbral de decisión, la observación se clasifica como perteneciente a la clase positiva, de lo contrario se clasifica como perteneciente a la clase negativa.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# Separamos datos de entrada y salida

x_df = df_train_e[
  ['building_id', 'geo_level_1_id', 'geo_level_2_id', 'geo_level_3_id',
  'count_floors_pre_eq', 'age', 'area_percentage', 'height_percentage',
  'land_surface_condition', 'foundation_type', 'roof_type',
  'ground_floor_type', 'other_floor_type', 'position',
  'plan_configuration', 'legal_ownership_status', 'count_families']
]
y_df = df_train_e['damage_grade']
```

```python
x_df_test = df_test_e[
  ['building_id', 'geo_level_1_id', 'geo_level_2_id', 'geo_level_3_id',
  'count_floors_pre_eq', 'age', 'area_percentage', 'height_percentage',
  'land_surface_condition', 'foundation_type', 'roof_type',
  'ground_floor_type', 'other_floor_type', 'position',
  'plan_configuration', 'legal_ownership_status', 'count_families']
]
```

```python
# Separamos train y test

x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.2, random_state=42)
```

#### Prueba de hiperparámetros

```python
param_dist = {
  "penalty": ['l2', None],
  "solver": ['lbfgs', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
  "n_jobs": [-1]
}

# RandomSearch
lr = RandomizedSearchCV(LogisticRegression(), param_dist, random_state=0)
search = lr.fit(x_train, y_train)
search.best_params_
```

**RandomSearch** es un algoritmo de búsqueda de hiperparámetros utilizado en el aprendizaje automático para encontrar la mejor combinación de parámetros de un modelo que maximice su rendimiento en un conjunto de datos.

El uso de RandomSearch puede ser beneficioso porque puede reducir significativamente el tiempo y el esfuerzo necesarios para encontrar los mejores parámetros para un modelo. Además, puede evitar el problema de sobreajuste al no limitarse a un conjunto fijo de parámetros. En cambio, puede explorar una gama más amplia de combinaciones de parámetros, lo que puede mejorar la precisión y el rendimiento general del modelo.

#### Entrenamiento

```python
lr = LogisticRegression(solver='newton-cholesky', penalty=None, n_jobs= -1, random_state=0)
lr.fit(x_train, y_train)
lr.coef_
```


El coeficiente (**coef**) en LogisticRegression se refiere a los coeficientes de la ecuación del modelo de regresión logística. Estos coeficientes son los valores que se ajustan al conjunto de datos de entrenamiento durante el proceso de entrenamiento del modelo para crear un modelo predictivo.

En la regresión logística, los coeficientes son utilizados para calcular la probabilidad de que un evento ocurra, como la probabilidad de que un correo electrónico sea spam o no spam. Cada variable de entrada en el modelo se multiplica por su correspondiente coeficiente y se suman todos los resultados para producir una estimación de la probabilidad del evento.

#### Métricas de Error

```python
y_pred = lr.predict(x_test)
print('Test F1 micro:', f1_score(y_pred, y_test, average='micro'))
```

```python
y_pred
```

```python
# Reporte

print(classification_report(y_pred, y_test, labels=[1, 2, 3]))
```

El **classification_report** es una función en la biblioteca Scikit-learn de Python que proporciona un resumen de los resultados de la clasificación de un modelo de aprendizaje automático. Este informe es especialmente útil para evaluar la precisión y el rendimiento de un modelo de clasificación.

El informe generado por la función classification_report muestra la precisión, el recall, la puntuación F1 y el soporte para cada clase en el conjunto de datos. La precisión representa la proporción de predicciones positivas correctas en relación con el número total de predicciones positivas realizadas por el modelo. El recall, por otro lado, representa la proporción de casos positivos identificados correctamente por el modelo en relación con el número total de casos positivos en los datos de prueba. La puntuación F1 es una medida de la precisión y el recall que se combina en una sola medida.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Matriz de confusión
cm = confusion_matrix(y_pred, y_test)
print(cm)
display_cm = ConfusionMatrixDisplay(cm, display_labels=[1,2,3])
fig, ax = plt.subplots(figsize=(8,8))
display_cm.plot(ax=ax)
plt.show()
```

La **matriz de confusión**, es una herramienta útil en el aprendizaje automático para evaluar el rendimiento de un modelo de clasificación. La matriz de confusión es una tabla que muestra la frecuencia con la que el modelo clasifica los datos correctamente y los errores que comete.

La matriz de confusión se presenta en una tabla en la que los valores verdaderos se sitúan en la diagonal principal, mientras que los falsos positivos y falsos negativos se sitúan en las otras dos celdas de la fila y columna correspondientes. La matriz de confusión proporciona una descripción detallada del rendimiento del modelo de clasificación y puede utilizarse para calcular una serie de métricas de evaluación, como la precisión, la sensibilidad, la especificidad y la tasa de error.

## Modelo 2: SVM

**Support Vector Classifier (SVC)** es un modelo de clasificación utilizado en el aprendizaje automático. Se basa en el concepto de máquinas de vectores de soporte (SVM) y es muy eficaz en la clasificación de datos en clases binarias o multiclase.

Es un modelo de aprendizaje supervisado, lo que significa que se entrena con datos etiquetados y luego se utiliza para clasificar datos no etiquetados. Durante el entrenamiento, el modelo ajusta los parámetros para maximizar la precisión de la clasificación.

También puede utilizarse en conjunción con técnicas de kernel, que permiten la clasificación no lineal de datos. Los kernels transforman los datos en un espacio de mayor dimensión donde los datos pueden ser más fácilmente separados por un hiperplano.
